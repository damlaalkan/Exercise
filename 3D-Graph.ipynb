```python
import vtk
import csv

class Tube:
    
    def __init__(self):        
        self.LineActors = []
        self.TubeActors = []    
        
    def CreateOneSegment(self, p1, p2,radius):    
        
            lineSource = vtk.vtkLineSource()        
            lineSource.SetPoint1(p1)        
            lineSource.SetPoint2(p2)   
            lineMapper = vtk.vtkPolyDataMapper()        
            lineMapper.SetInputConnection(lineSource.GetOutputPort())        
            lineActor = vtk.vtkActor()        
            lineActor.GetProperty().SetColor(100.0, 700.0, 0.0)        
            lineActor.SetMapper(lineMapper)        
            tubeFilter = vtk.vtkTubeFilter()        
            tubeFilter.SetInputConnection(lineSource.GetOutputPort())        
            tubeFilter.SetRadius(radius)        
            tubeFilter.SetNumberOfSides(50)        
            tubeFilter.Update()        
            tubeMapper = vtk.vtkPolyDataMapper()        
            tubeMapper.SetInputConnection(tubeFilter.GetOutputPort())        
            tubeActor = vtk.vtkActor()        
            tubeActor.GetProperty().SetOpacity(0.5)        
            tubeActor.SetMapper(tubeMapper)      
            
            self.LineActors.append(lineActor)        
            self.TubeActors.append(tubeActor)   
            
            
    def Show(self):
        
        renderer = vtk.vtkRenderer()   
        
        for (la, ta) in zip(self.LineActors, self.TubeActors):            
            renderer.AddActor(la)            
            renderer.AddActor(ta)   
            
        rwindow = vtk.vtkRenderWindow()        
        rwindow.AddRenderer(renderer)  
        
        rwindow_interactor = vtk.vtkRenderWindowInteractor()        
        rwindow_interactor.SetRenderWindow(rwindow) 
        
        rwindow.Render()        
        rwindow_interactor.Start()
        
        
if __name__ == '__main__':
    
   with open('tube_points.csv') as csvfile:
       
       readCSV=csv.reader(csvfile,delimiter=',')
       
       radius=[]
       points=[]
       
       for row in readCSV:
           if row[0]== 'p' :
              
               r= float(row[4])
               radius.append(r)
               
               x,y,z = int(row[1]),int(row[2]),int(row[3])
               points.append((x,y,z))
           else:
            pass
       

t=Tube()
for i in range(len(points)-1):
    t.CreateOneSegment(points[i],points[i+1],radius[i])
t.Show()

```
